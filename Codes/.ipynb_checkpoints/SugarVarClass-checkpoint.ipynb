```python
#Import Libraries

import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

from skimage.feature import hog
```

```python
# Define dataset paths
base_path = "/content/drive/My Drive/Research/DatasetAugmented_Split"
train_path = os.path.join(base_path, "train")
val_path = os.path.join(base_path, "val")
test_path = os.path.join(base_path, "test")

# Define class labels
classes = ["v1", "v2", "v3", "v4"]
IMG_SIZE = (128, 128)  # Resize images
```

```python
# Function to load images and extract features
def extract_features(img_path):
    img = cv2.imread(img_path)
    img = cv2.resize(img, IMG_SIZE)  # Resize
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Convert BGR to RGB

    # Histogram Features (Color Distribution)
    hist_r = cv2.calcHist([img], [0], None, [256], [0, 256])
    hist_g = cv2.calcHist([img], [1], None, [256], [0, 256])
    hist_b = cv2.calcHist([img], [2], None, [256], [0, 256])

    hist_features = np.concatenate((hist_r.flatten(), hist_g.flatten(), hist_b.flatten()))

    # HOG Feature Extraction
    gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)  # Convert to grayscale
    hog_features = hog(gray, orientations=9, pixels_per_cell=(8, 8),
                       cells_per_block=(2, 2), block_norm='L2-Hys', feature_vector=True)

    # Combine HOG + Histogram features
    combined_features = np.concatenate((hist_features, hog_features))

    return combined_features  # New feature vector
```

```python
# Function to load dataset & extract features
def load_dataset(folder_path, label):
    features = []
    labels = []

    for filename in os.listdir(folder_path):
        img_path = os.path.join(folder_path, filename)
        if os.path.isfile(img_path):
            feature_vector = extract_features(img_path)
            features.append(feature_vector)
            labels.append(label)

    return features, labels
```

```python
#ensure the data
!echo "v1:" && find "/content/drive/My Drive/Research/DatasetAugmented_Split/train/v1" -type f | wc -l
```

```python
# Load all dataset folders
X_train, y_train = [], []
X_val, y_val = [], []
X_test, y_test = [], []

for idx, class_name in enumerate(classes):
    train_feats, train_labels = load_dataset(os.path.join(train_path, class_name), idx)
    val_feats, val_labels = load_dataset(os.path.join(val_path, class_name), idx)
    test_feats, test_labels = load_dataset(os.path.join(test_path, class_name), idx)

    X_train.extend(train_feats)
    y_train.extend(train_labels)
    X_val.extend(val_feats)
    y_val.extend(val_labels)
    X_test.extend(test_feats)
    y_test.extend(test_labels)
```

```python
# Convert to NumPy arrays
X_train, y_train = np.array(X_train), np.array(y_train)
X_val, y_val = np.array(X_val), np.array(y_val)
X_test, y_test = np.array(X_test), np.array(y_test)

print("Dataset Loaded!")
print(f"Train: {X_train.shape}, Val: {X_val.shape}, Test: {X_test.shape}")
```

**Train ML Models**

```python
#Train SVM
svm_model = SVC(kernel="linear")
svm_model.fit(X_train, y_train)
```

```python
#Train Random Forest
rf_model = RandomForestClassifier(n_estimators=300, max_depth=50, random_state=42)
rf_model.fit(X_train, y_train)
```

```python
# Evaluate Model
#y_pred_svm = svm_model.predict(X_test)
y_pred_rf = rf_model.predict(X_test)

# Print Accuracy
#print("\nSVM Accuracy:", accuracy_score(y_test, y_pred_svm))
print("Random Forest Accuracy:", accuracy_score(y_test, y_pred_rf))

# Print Classification Reports
#print("\nSVM Classification Report:\n", classification_report(y_test, y_pred_svm))
print("\nRandom Forest Classification Report:\n", classification_report(y_test, y_pred_rf))
```

```python
# Confusion Matrix
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.title("SVM Confusion Matrix")
plt.imshow(confusion_matrix(y_test, y_pred_svm), cmap="Blues", interpolation="nearest")

plt.subplot(1,2,2)
plt.title("Random Forest Confusion Matrix")
plt.imshow(confusion_matrix(y_test, y_pred_rf), cmap="Greens", interpolation="nearest")

plt.show()
```
